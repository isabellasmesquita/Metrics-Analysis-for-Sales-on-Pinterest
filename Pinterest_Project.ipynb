```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Load the dataset
df = pd.read_csv('C:/Users/isabe/OneDrive/Ambiente de Trabalho/project module 1/pinterest_finalised.csv')

# Display the column names
print(df.columns)

# Create the content_type column based on the description or title
def determine_content_type(row):
    # Check if the description is not NaN
    if isinstance(row['description'], str):  # Confirm it is a string
        if 'image' in row['description'].lower():
            return 'Image Only'
        elif 'video' in row['description'].lower():
            return 'Video Only'
    return 'Both'  # or another type you want to define when the condition is not met

# Apply the function to create the new column
df['content_type'] = df.apply(determine_content_type, axis=1)

# Calculate the average repins by content type
engagement_by_type = df.groupby('content_type')['repin_count'].mean().reset_index()

# Create the bar chart
plt.figure(figsize=(10, 6))

# Plot the data
sns.barplot(data=engagement_by_type, x='content_type', y='repin_count')

plt.title('Average Repins by Content Type')
plt.xlabel('Content Type')
plt.ylabel('Average Repins')

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from collections import Counter
import re

# Load the CSV file
df = pd.read_csv('pinterest_finalised.csv', encoding='MacRoman')

# Check the data
print("Total number of rows:", len(df))
print("\nNumber of non-null values in each column:")
print(df.notnull().sum())

# Function to clean the text
def clean_text(text):
    if isinstance(text, str):
        # Remove special characters and convert to lowercase
        return re.sub(r'[^\w\s]', '', text.lower())
    return ''

# Clean and count the words in the titles
words = df['title'].fillna('').apply(clean_text).str.split(expand=True).stack().value_counts()

print("\nThe 20 most common words in the titles:")
print(words.head(20))

# Plot the 20 most common words
plt.figure(figsize=(12, 6))
words.head(20).plot(kind='bar')
plt.title('20 Most Common Words in Pinterest Titles')
plt.xlabel('Words')
plt.ylabel('Frequency')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from collections import defaultdict
import re

# Load the CSV file
df = pd.read_csv('pinterest_finalised.csv', encoding='MacRoman')

# Function to clean the text
def clean_text(text):
    if isinstance(text, str):
        return re.sub(r'[^\w\s]', '', text.lower()).split()
    return []

# Create a dictionary to store the repin count for each word
word_repin_count = defaultdict(list)

# Iterate over the rows of the DataFrame
for _, row in df.iterrows():
    title = row['title']
    repin_count = row['repin_count']
    
    if pd.notna(title):
        words = clean_text(title)
        for word in words:
            word_repin_count[word].append(repin_count)

# Calculate the average repins for each word
word_avg_repins = {word: sum(counts) / len(counts) for word, counts in word_repin_count.items() if len(counts) > 10}

# Sort words by average repins
sorted_words = sorted(word_avg_repins.items(), key=lambda x: x[1], reverse=True)

# Take the top 20 words
top_20_words = sorted_words[:20]

# Create separate lists for words and average repins
words = [word for word, _ in top_20_words]
avg_repins = [avg for _, avg in top_20_words]

# Create the bar plot
plt.figure(figsize=(12, 6))
sns.barplot(x=words, y=avg_repins)
plt.title('Top 20 Words by Average Repins')
plt.xlabel('Words')
plt.ylabel('Average Repins')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()

# Print the results
print("Top 20 words by average repins:")
for word, avg in top_20_words:
    print(f"{word}: {avg:.2f}")

# Calculate the correlation between title length and repin count
df['title_length'] = df['title'].fillna('').apply(len)
correlation = df['title_length'].corr(df['repin_count'])

print(f"\nCorrelation between title length and number of repins: {correlation:.2f}")

# Create a scatter plot
plt.figure(figsize=(10, 6))
plt.scatter(df['title_length'], df['repin_count'], alpha=0.5)
plt.title('Relationship Between Title Length and Number of Repins')
plt.xlabel('Title Length')
plt.ylabel('Number of Repins')
plt.show()


```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from collections import Counter

# Load the data
qa = pd.read_csv('pinterest_Q_A.csv', encoding='ascii')

# Function to get all unique keywords
def get_all_keywords(series):
    return [keyword.strip() for keywords in series.dropna() for keyword in keywords.split(',')]

# Get all keywords
all_keywords = get_all_keywords(qa['keywords'])

# Count keyword frequencies
keyword_counts = Counter(all_keywords)

# Get top 15 keywords
top_15_keywords = dict(keyword_counts.most_common(15))

# Create a horizontal bar plot
plt.figure(figsize=(12, 8))
sns.barplot(x=list(top_15_keywords.values()), y=list(top_15_keywords.keys()), palette='viridis')
plt.title('Top 15 Keywords - Horizontal Bar Plot')
plt.xlabel('Frequency')
plt.ylabel('Keywords')
plt.show()
```
